```python
mypro=mypro[["TITLE","DESC","MRP","categories","Category-2","Category-3","Category-4","BRAND","COLOR","DETAILED DESC"]]
```

```python
# take dataset of size 10000 records for redusing computation power 
mypro=mypro.iloc[:10000]
```

```python
mypro.insert(0, 'id', range(0,0 + len(mypro)))
mypro
```

```python
mydf=mypro[["DETAILED DESC","id"]]
```

```python
mydf.dropna(axis=0,how="any",inplace=True)
```

```python
mydf.shape
```

```python
# import tfidfvectorizer 
from sklearn.feature_extraction.text import TfidfVectorizer
# define TF IDF Vestorizer object,remove all english stopword
tfidf=TfidfVectorizer(stop_words="english")
mydf["DETAILED DESC"]=mydf["DETAILED DESC"].fillna(' ')
# constructing IF-IDF Matrix by applying fit_transform mathed on overview feature
tfidf_matrix=tfidf.fit_transform(mydf["DETAILED DESC"])
```

```python
# import linear_kernal to calculate dot product
from sklearn.metrics.pairwise import linear_kernel
# compute the cosine similarity matrix
cosine_sim=linear_kernel(tfidf_matrix,tfidf_matrix)
```

```python
cosine_sim
```

```python
indices = pd.Series(mypro.index, index=mypro['TITLE']).drop_duplicates()
```

```python
indices.head()
```

```python
# Function that takes in movie title as input and gives recommendations 
def content_recommender(TITLE, cosine_sim=cosine_sim, mypro=mypro, indices=indices):
    # Obtain the index of the movie that matches the title
    idx = indices[TITLE]

    # Get the pairwsie similarity scores of all movies with that movie
    # And convert it into a list of tuples as described above
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    # Sort the movies based on the cosine similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 10 most similar movies. Ignore the first movie.
    sim_scores = sim_scores[1:11]

    # Get the movie indices
    movie_indices = [i[0] for i in sim_scores]

    # Return the top 10 most similar movies
    return mypro['TITLE'].iloc[movie_indices]
```

```python
content_recommender("Dermawear Men's Shapewear")
```
